```python
# Import libraries
import numpy as np
import pandas as pd
import datetime as datetime
import matplotlib.pyplot as plt
from plotnine import ggplot, aes, geom_line, labs
```

```python
# Import data sets
df_canada = pd.read_excel('search_term_canada.xlsx')
#df_iran = pd.read_excel('search_term_iran.xlsx')
df_italy = pd.read_excel('search_term_italy.xlsx')
df_japan = pd.read_excel('search_term_japan.xlsx')
df_sk = pd.read_excel('search_term_sk.xlsx')
df_uk = pd.read_excel('search_term_uk.xlsx')
df_us = pd.read_excel('search_term_us.xlsx')
df_world = pd.read_excel('search_term_worldwide.xlsx')
```

```python
# Check world dataset
df_world.head()
```

```python
# Set up next few visualizations
df_world['Week'] = pd.to_datetime(df_world['Week'])
df_world = df_world[(df_world['Week'] >= '2019-12-01')]

dfs = [df_canada, df_italy, df_japan, df_sk, df_uk, df_us]
countries = ['Canada', 'Italy', 'Japan', 'South Korea', 'United Kingdom', 'United States']
colors = ['lightsteelblue', 'lightgreen', 'lightskyblue', 'lightpink', 'lightseagreen', 'lightsalmon']
```

## WORLD PLOTS

### Mental Health

```python
# Plot OF MENTAL HEALTH
plt.figure(figsize=(12, 6)) 
plt.plot(df_world['Week'], df_world['mental health'], marker='x', color='lightsteelblue', linestyle='-')
plt.title("Timeline of Search Interest on Google of 'Mental Health' Worldwide")  # Including 'Mental Health' within the title
plt.xlabel('Week')
plt.ylabel("Search Interest of 'Mental Health' Worldwide")
plt.grid(True)
plt.xticks(pd.date_range(start=df_world['Week'].iloc[0], end=df_world['Week'].iloc[-1], freq='MS'), rotation=45)
plt.tight_layout()
plt.savefig('mental_health_timeline_world.png')
```

## COUNTRIES

### Mental Health

```python
# Preprocess the data

plt.figure(figsize=(12, 6)) 
# Plot each country's data
for i, df in enumerate(dfs):
    df['Week'] = pd.to_datetime(df['Week'])
    df = df[(df['Week'] >= '2019-12-01')]
    plt.plot(df['Week'], df['mental health'], label=countries[i], color=colors[i], linestyle='-', marker='x')

plt.title("Timeline of Search Interest on Google for 'Mental Health' by Country")
plt.xlabel('Week')
plt.ylabel("Search Interest of 'Mental Health'")
plt.grid(True)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.savefig('mental_health_timeline_overlay.png')  
plt.show()
```

### Obsessive Compulsive Disorder

```python
plt.figure(figsize=(12, 6)) 
# Plot each country's data
for i, df in enumerate(dfs):
    df['Week'] = pd.to_datetime(df['Week'])
    df = df[(df['Week'] >= '2019-12-01')]
    
    # Calculate the added values between consecutive values
    added_values = df['obsessive compulsive disorder'] + df['obsessive compulsive disorder'].shift(-1)
    
    # Plot the average values
    plt.plot(df['Week'][:-1], added_values[:-1], label=countries[i], color=colors[i], linestyle='-', marker='x')

plt.title("Timeline of Search Interest on Google for 'OCD' by Country")
plt.xlabel('Week')
plt.ylabel("Search Interest of 'OCD or Obsessive Compulsive Disorder'")
plt.grid(True)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.savefig('OCD_timeline_overlay.png')  
plt.show()
```
